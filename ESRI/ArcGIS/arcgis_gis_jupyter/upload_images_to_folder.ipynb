```python
# UPLOAD_IMAGES_TO_PORTAL
# uploads images in the same directory as this script to target portal 'WebAppImages' folder, which must be created prior to running
# arcgis version 2.3.1, Python version 3.12.3

### CONFIGURATION USERNAMES AND PASSWORDS
from arcgis.gis import *
import json, requests, urllib, os, datetime
from pathlib import Path
from arcgis.gis._impl._content_manager import SharingLevel

workspace =  os.path.dirname(os.path.abspath("__file__"))

env_prefixes = {
    "GLOBAL_DEV": {
        "server": "dev.domain.com",
        "wa": "server",
        "folder": "ATLAS",
        "portal_url": "https://dev.domain.com/portal",
        "portal_user": os.environ["GLOBAL_DEV_PORTAL_USER"],
        "portal_pw": os.environ["GLOBAL_DEV_PORTAL_PW"]
    }
}

portal_img_folder = 'WebAppImages'

def upload_image(gis, img_path, portal_img_folder):
    try:
        item_properties = {'type':'Image'}
        item_properties['title'] = img_path.stem
        
        new_item = gis.content.add(
            item_properties=item_properties,
            data=str(img_path),
            folder=portal_img_folder
        )
        
        print('uploaded {}({})'.format(img_path, new_item.itemid))

        # set the sharing level to public
        new_item.sharing.sharing_level = SharingLevel.EVERYONE
        
    except:
        return None
    

for env_key in env_prefixes:
    gis = GIS(url=env_prefixes[env_key]['portal_url'], username=env_prefixes[env_key]['portal_user'], password=env_prefixes[env_key]['portal_pw'])
    
    user = gis.users.me
    
    item_properties = {
        "type": "Image"
    }
    

    item_ids = []
    
    for img_path in Path(workspace).glob('*.jpeg'):
        upload_image(gis, img_path, portal_img_folder)
    
    for img_path in Path(workspace).glob('*.jpg'):
        upload_image(gis, img_path, portal_img_folder)
        
    for img_path in Path(workspace).glob('*.png'):
        upload_image(gis, img_path, portal_img_folder)
```
